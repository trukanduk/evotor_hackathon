```python
import pandas as pd
import numpy as np
```

```python
onhand = pd.read_csv("onhand_table.csv")
```

```python
N_s = [17846, 23293, 10171, 11013, 58864, 108664]
filenames = ["kofe.csv", "pivo.csv", "alco.csv", "prod.csv", "meat.csv", "larek.csv"]
```

```python
onhand
```

```python
for N, filename in zip(N_s, filenames):
    df = pd.read_csv(filename)
    temp = onhand[onhand["shop_id"] == N].values
    good_to_onhand = dict(zip(temp[:, 1], temp[:, 3]))
    new_column = []
    for good in df["bar_code"].values:
        if good in good_to_onhand:
            new_column += [good_to_onhand[good]]
        else:
            new_column += [0]
    df["count"] = np.array(new_column)
    df.to_csv("new" + filename, index=False)
```
